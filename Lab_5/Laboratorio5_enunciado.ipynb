### 1.3.4 Información por Categorías [1 punto]

````python
df_grades = pd.concat(
    [
        df_grades_1, 
        df_grades_2
    ]
    , axis=0)

# Primero calculamos el promedio
def alumnos_promedio(df):
    df["GPA"] = (df["math score"] + df["reading score"] + df["writing score"])/3
    df["GPA"] = df["GPA"].round(2)
    return df
df_grades = alumnos_promedio(df_grades)
df_grades.head(4)
````

````python
mapeo = {'some high school':'school','some college':'school','high school':'school',"bachelor's degree":'college',
         "associate's degree":"college","master's degree": "postgraduate"}

df_grades['simple parental level of education'] = df_grades['parental level of education'].map(mapeo)
df_grades.head()
````

````python
mean = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])['math score', 'reading score', 'writing score','GPA'].mean().round(2).reset_index()
count = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')

n = df_grades.shape[0]
count['percentage'] = (count['count'] / n*100).round(2).astype(str) + ' %'
result = pd.merge(mean, count, on=['race/ethnicity', 'simple parental level of education'], how='inner')
result.head(15)
````

````python
new_df = pd.read_csv('other_grades.csv')
df_outer_join = pd.merge(df_grades, new_df, on='names', how='outer')
print(df_outer_join.shape)
df_left_join = pd.merge(df_grades, new_df, on='names', how='left')
print(df_left_join.shape)
df_right_join = pd.merge(df_grades, new_df, on='names', how='right')
print(df_right_join.shape)
df_inner_join = pd.merge(df_grades, new_df, on='names', how='inner')
print(df_inner_join.shape)
````

> **Respuesta**: 

- Outer: Este combina todos los elementos posibles (es decir, todos los que están en df_grades y todos los que están en other_grades.csv). Por lo tanto, contiene todos los que ya teniamos en df_grades y además agrega todos los que ya habían sido filtrados anteriormente, y que están ahora solo en other_grades.csv. (CONTIENE LOS DATOS LIMPIOS Y LOS QUE HABÍAN SIDO ELIMINADOS EN LA LIMPIEZA)

- Left: Este tiene todos los elementos que ya estaban en df_grades, y además agrega la info de los que están en ambos df. (Es decir, este cuenta con la información de las nuevas dos materias de todas las filas que ya estaban en df_grades, siendo estos los datos que ya habían sido filtrados)

- Right: Este conserva solo los elementosque se hayan combinado correctamente provenientes del dataset right. (Es decir, contiene los datos que pasaron por la limpieza y quedaron, así como los que no quedaron finalmente, siendo estos los mismos que existen en el outer join) 

- Inner: Este combina solo los elementos que se encuentran en las dos tablas, y elimina todo el resto. (Es decir, tiene solo los que ya estaban en df_grades y por lo tanto, es igual a el left join. (EN ESTE CASO))

Nota: Asumiendo que el primer dataframe está correcto, lo mejor sería utilizar "left join", ya que en la salida se mantendrían los alumnos que están presentes en el primer dataframe. Para un caso más general se buscaría que los alumnos tengan todas sus notas, por ende habría que utilizar "inner join", buscando que en el resultado final esten los alumnos que tienen todas sus notas.
