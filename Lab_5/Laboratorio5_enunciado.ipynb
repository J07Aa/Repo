### 1.3.4 Informaci칩n por Categor칤as [1 punto]

````python
df_grades = pd.concat(
    [
        df_grades_1, 
        df_grades_2
    ]
    , axis=0)

# Primero calculamos el promedio
def alumnos_promedio(df):
    df["GPA"] = (df["math score"] + df["reading score"] + df["writing score"])/3
    df["GPA"] = df["GPA"].round(2)
    return df
df_grades = alumnos_promedio(df_grades)
df_grades.head(4)
````

````python
mapeo = {'some high school':'school','some college':'school','high school':'school',"bachelor's degree":'college',
         "associate's degree":"college","master's degree": "postgraduate"}

df_grades['simple parental level of education'] = df_grades['parental level of education'].map(mapeo)
df_grades.head()
````

> **Respuesta**: 

- Outer: Este combina todos los elementos posibles (es decir, todos los que est치n en df_grades y todos los que est치n en other_grades.csv). Por lo tanto, contiene todos los que ya teniamos en df_grades y adem치s agrega todos los que ya hab칤an sido filtrados anteriormente, y que est치n ahora solo en other_grades.csv. (CONTIENE LOS DATOS LIMPIOS Y LOS QUE HAB칈AN SIDO ELIMINADOS EN LA LIMPIEZA)

- Left: Este tiene todos los elementos que ya estaban en df_grades, y adem치s agrega la info de los que est치n en ambos df. (Es decir, este cuenta con la informaci칩n de las nuevas dos materias de todas las filas que ya estaban en df_grades, siendo estos los datos que ya hab칤an sido filtrados)

- Right: Este conserva solo los elementosque se hayan combinado correctamente provenientes del dataset right. (Es decir, contiene los datos que pasaron por la limpieza y quedaron, as칤 como los que no quedaron finalmente, siendo estos los mismos que existen en el outer join) 

- Inner: Este combina solo los elementos que se encuentran en las dos tablas, y elimina todo el resto. (Es decir, tiene solo los que ya estaban en df_grades y por lo tanto, es igual a el left join. (EN ESTE CASO))

Nota: Asumiendo que el primer dataframe est치 correcto, lo mejor ser칤a utilizar "left join", ya que en la salida se mantendr칤an los alumnos que est치n presentes en el primer dataframe. Para un caso m치s general se buscar칤a que los alumnos tengan todas sus notas, por ende habr칤a que utilizar "inner join", buscando que en el resultado final esten los alumnos que tienen todas sus notas.
